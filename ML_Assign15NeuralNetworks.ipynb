# Experiment 5: Overfitting Detective

**Goal**: Create a model that overfits (memorizes instead of learning).

**Steps**: Pick the Gaussian dataset. Add noise (50). Add 4+ hidden layers with 8 neurons each. Set regularization to None. Train for 1000+ epochs.

**What to watch**: Does the training loss go down while the test loss stays high or goes up? That's overfitting!

**Now try**: Turn on L2 regularization and retrain. Did it help?

**Answer:** When a model overfits, it works excellently on training data but fails during new data. It overlooks underlying patterns and prioritizes memorizing. Turning on L2 regularization fixes this by penalizing large weights, forcing the model to find a simpler solution.

**1. Circle Dataset with No Hidden Layers**

A network with no layers can only draw a straight line to separate data.. The circle dataset needs a curved boundary, which a straight line simply can't do. So the model fails because its not complex enough to solve the circle dataset.

**2. Best Learning Rate for Gaussian Dataset **

A medium learning rate works for Gaussian data. If the learning rate is too high the model overshoots the answer. Bounces around without settling. If its too low it learns slowly it barely improves. A learning rate of around 0.01 to 0.03 usually works best for the dataset.

**3. Minimum Neurons/Layers for XOR**

The XOR problem needs least 1 hidden layer with 2 neurons to solve it. Simpler networks can't separate XOR data linearly. The hidden layer helps the network combine inputs in a -linear way to solve the XOR problem.

**4. Biggest Fix in Experiment 4**

Adding a hidden layer makes the difference. It gives the model the ability to learn -linear patterns in the data. Changing activation or tweaking the learning rate helps,. Without depth the model is limited. The structure of the model matters more than tuning.

**5. Overfitting vs. Underfitting by Loss**

Underfitting happens when both training loss and test loss are high. This means the model isn't learning much all from the training data. Overfitting happens when training loss is low but test loss is high. The model has memorized the training data. Fails on new data. The gap between the two losses is a warning sign for overfitting.
