##Q1) What is Density Based Clustering?

Density-based clustering is a way to group data points together based on how closely packed they are in space. The algorithm groups those points which are densely packed (have lots of neighbours) and considers the isolated ones outliers.

##Q2) What are its applications?

-- Urban Planning: Identifying high population density, traffic congestion clusters, etc

-- Market Analysis: Identifying patterns of similar product purchases or pricing ranges

-- Fraud Detection: Outliers in banking or credit data can be identified as scams

-- Recommendations: Grouping posts, reels, etc of similar content

##Q3) Why is it better/worse than K-Means Clustering?

I would say Density based Cluttering is better than K Means Clustering as in the latter you have to declare the number of clusters beforehand. It is much more flexible and automatically discovers the natural patterns of the data without having to predetermine anything. Though K Means is faster, but it has too many rigid assumptions.

## Q4. Illustrate Density Based Clustering on a data set you create from scrath.

```python
#@title Q4

from sklearn.cluster import DBSCAN
import numpy as np

# Sample data
dataset = np.array([[2, 2], [2.5, 2.3], [2.1, 2.4],
                   [8, 8], [8.3, 8.1], [8.1, 8.4],
                   [5, 5], [12, 3]])

# Create DBSCAN model
dbscan = DBSCAN(
    eps=1.0,        # Neighborhood radius
    min_samples=2   # Minimum points for core
)

# Fit and predict
labels = dbscan.fit_predict(dataset)
print(f"Cluster labels: {labels}")
print(f"Noise points: {list(labels).count(-1)}")
print(f"Number of clusters: {len(set(labels)) - (1 if -1 in labels else 0)}")
```

##Q5:  Illustrate Density Based Clustering on a toy data set from the scikit-learn

```python
#@title Q5
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import numpy as np

# Generate make_moons dataset from scikit-learn
dataset, _ = make_moons(n_samples=300, noise=0.05, random_state=42)

# Create DBSCAN model
dbscan = DBSCAN(
    eps=1.0,           # Neighborhood radius
    min_samples=2      # Minimum points for core
)

# Fit and predict
labels = dbscan.fit_predict(dataset)
print(f"Cluster labels: {labels}")
print(f"Noise points: {list(labels).count(-1)}")
print(f"Number of clusters: {len(set(labels)) - (1 if -1 in labels else 0)}")
```

SUPERVISED EVALUATION

-- NMI: Measures info shared between clusters, between 0 to 1, where 1 is perfect
-- Fowlkes-Mallows Index: Recall for pairs of points
